Q1. Explain GET and POST methods.

GET: GET is a method used to retrieve data from a server. When a client makes a GET request, it asks the server to return a specific resource identified by a URL (Uniform Resource Locator). The data is appended to the URL as query parameters. For example, if you enter a URL in your web browser and press Enter, you're typically making a GET request.

POST: POST is a method used to submit data to the server to create or update a resource. When a client makes a POST request, it sends the data in the body of the HTTP request. This allows for sending larger amounts of data compared to GET requests, and the data is not directly visible in the URL.


Q2. Why is request used in Flask?

In Flask, the request object is used to handle incoming HTTP requests from clients. It provides access to various aspects of the request, such as form data, query parameters, headers, cookies, and the request body. The request object is an instance of the Request class, which encapsulates all the information related to the current request.

The request object in Flask is essential because it allows you to interact with the data sent by the client and perform actions based on that data.

Q3. Why is redirect() used in Flask?

A redirect is used in the Flask class to send the user to a particular URL with the status code. conversely, this status code additionally identifies the issue. When we access a website, our browser sends a request to the server, and the server replies with what is known as the HTTP status code, which is a three-digit number.

eg- 302 - found

    305 - Use_proxy
    
    306- Reserved   
    
    etc

```python
from flask import Flask,redirect

app = Flask(__name__)

@app.route("/")
def home():
    return redirect("/helloworld")


@app.route("/helloworld")
def hello_world():
    return "Hello, World from redirected page.!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q4. What are templates in Flask? Why is the render_template() function used?

Templates - In Flask, templates are used to generate dynamic HTML or other types of content that can be sent back to the client as a response. Templates allow you to separate the presentation logic from the application logic, making it easier to design and maintain your web pages.

render_template() - In Flask, the render_template function is a built-in function that is used to render a template and generate the final HTML or other content that will be sent as a response to the client's request. The render_template function is provided by the Flask framework and uses the Jinja2 templating engine to process the templates.

Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask,request , jsonify

app = Flask(__name__)


@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
