Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library for Python. It provides a wide range of tools and functions for creating various types of plots and charts, allowing users to visualize and explore their data effectively.

Uses:-

1. Data Visualization: Matplotlib enables the creation of high-quality, publication-ready visualizations. It offers a flexible and comprehensive set of plotting functions, allowing users to create a wide range of charts, graphs, and plots.

2. Customization and Styling: Matplotlib provides extensive options for customizing plots. Users can control the appearance of elements such as colors, markers, line styles, labels, axes, and annotations. This allows for fine-tuning and creating visually appealing and informative visualizations.

3. Wide Range of Plots: Matplotlib's pyplot module provides a simple and convenient interface to create various types of plots. Some commonly used plots include line plots, scatter plots, bar plots, histogram plots, and pie charts.

4. Integration with NumPy and Pandas: Matplotlib seamlessly integrates with other popular Python libraries, such as NumPy and Pandas, making it easy to plot data stored in arrays, Series, or DataFrames.

5 plots are:-

1. Line Plot: A line plot displays data points connected by straight line segments. It is useful for visualizing trends or patterns in data over a continuous variable.

2. Scatter Plot: A scatter plot represents individual data points as markers on a 2D plane. It is often used to show the relationship between two variables and identify any patterns or clusters in the data.

3. Bar Plot: A bar plot, also known as a bar chart, displays data using rectangular bars of varying heights. It is commonly used to compare categorical data or display counts or frequencies.

4. Histogram: A histogram is a graphical representation of the distribution of a dataset. It divides the data into bins and displays the frequency or count of values falling within each bin.

5. Pie Chart: A pie chart represents data as slices of a circular pie, with each slice representing a different category. It is useful for illustrating proportions or percentages of a whole.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that represents individual data points as markers on a two-dimensional coordinate system. It is particularly useful for displaying the relationship between two numerical variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatter plot")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the plotting area into multiple smaller plots or subplots, each with its own set of axes.

The subplot() function takes three arguments: num_rows, num_cols, and plot_number. These arguments define the grid structure and specify which subplot you want to create or modify.

few reasons, the subplot() function is useful:

1. Multiple Plots in One Figure: With subplot(), you can create multiple subplots within a single figure, allowing you to visualize and compare different aspects of your data side by side. This is particularly useful when you want to display related plots together or present different views of the same data.

2. Grid Layout Control: The subplot() function provides control over the layout of subplots using the num_rows and num_cols arguments. These arguments define the number of rows and columns in the subplot grid, respectively. You can specify any configuration to suit your plotting needs.

3. Easy Navigation and Plot Selection: The plot_number argument of subplot() allows you to select a specific subplot within the grid. By specifying the row and column position of a subplot, you can easily navigate to a particular subplot and modify its properties, such as data, axes labels, title, etc.

4. Flexible Plot Arrangements: The subplot() function enables you to arrange subplots in various configurations, including grids of equal-sized subplots or custom-sized subplots. You can adjust the size and position of each subplot within the grid, allowing for different visualizations of your data.

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)

plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)

plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)

plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)

plt.plot(x,y)

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart, is a type of data visualization that displays categorical data with rectangular bars. Each bar represents a category, and the length or height of the bar corresponds to the value or frequency of that category.

Bar plots are commonly used to:

1. Compare Categories: Bar plots provide a visual comparison between different categories or groups. The length or height of each bar makes it easy to compare the values of different categories and identify the largest or smallest values.

2. Display Frequencies or Counts: Bar plots are useful for representing frequencies or counts of categorical data. The height of each bar represents the frequency or count of the corresponding category, allowing for a clear understanding of the distribution of the data.

3. Show Patterns or Trends: Bar plots can reveal patterns or trends across different categories. By arranging the bars in a specific order, such as ascending or descending, you can identify patterns and understand how different categories compare to each other.

4. Present Categorical Data: Bar plots are effective for visualizing categorical data where the x-axis represents the categories and the y-axis represents the values or frequencies. They are particularly useful when dealing with non-numerical or discrete data.

5. Bar plots can be created using various tools and libraries, such as Matplotlib, Seaborn, and Pandas. They offer customization options to control the appearance of bars, labels, axes, and other elements to create informative and visually appealing visualizations.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('Bar Plot')
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel('profit')
plt.ylabel('company')
plt.title('Horizontal Bar Plot')

plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution and summary statistics of a dataset. It provides a concise and visual representation of the minimum, first quartile, median, third quartile, and maximum values, as well as any outliers present in the data.

Box plots are commonly used to:

1. Show Data Distribution: Box plots offer a graphical summary of the distribution of a dataset. The box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The median is represented by a line inside the box, and the whiskers extend from the box to the minimum and maximum values (excluding outliers). This allows for a quick assessment of the spread and skewness of the data.

2. Identify Outliers: Box plots provide a visual indication of any outliers present in the dataset. Outliers are displayed as individual points or dots outside the whiskers, helping to identify data points that deviate significantly from the rest of the data.

3. Compare Distributions: Box plots are useful for comparing distributions across different groups or categories. By creating separate box plots for each group and placing them side by side, you can easily compare their medians, quartiles, and spread. This helps in identifying differences or similarities between groups.

4. Detect Skewness and Symmetry: Box plots provide insights into the skewness and symmetry of the data distribution. Skewed distributions can be identified by observing the position of the median within the box. If the median is closer to one end of the box than the other, it indicates skewness in the data

```python
box1 = np.random.normal(100, 10, 200)

plt.boxplot(box1)
plt.show()
```

```python
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box2)

plt.show()
```
