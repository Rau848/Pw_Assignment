Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful data visualization library in Python that builds on top of Matplotlib and provides additional functionality and aesthetic enhancements. Here are five commonly used plots that can be created using Seaborn, along with their typical uses:

1. Scatter Plot (scatterplot):

Use: Scatter plots are used to visualize the relationship between two numerical variables. They help identify patterns, correlations, and clusters in the data.

2. Line Plot (lineplot):

Use: Line plots are used to show the trend or change in a numerical variable over a continuous variable, such as time. They are effective for visualizing time series data or trends over a continuous x-axis.

3. Bar Plot (barplot):

Use: Bar plots are used to compare categories or groups. They show the distribution or comparison of numerical values across different categories. They are useful for displaying counts, frequencies, or aggregated data.

4. Histogram (histplot):

Use: Histograms are used to visualize the distribution of a numerical variable. They show the frequency or count of values falling within different bins or intervals. Histograms help identify patterns like skewness, multimodality, or outliers.

5. Heatmap (heatmap):

Use: Heatmaps are used to visualize the correlation or relationship between multiple variables in a tabular format. They use color encoding to represent the strength or magnitude of the relationship, making it easy to identify patterns or clusters in the data.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri,hue='event',style='region')
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamond=sns.load_dataset('diamonds')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flight=sns.load_dataset('flights')
```

```python
flight
```

```python
flights_pivot = flight.pivot(index='year', columns='month', values='passengers')
sns.heatmap(flights_pivot, cbar=False, linewidths=0.5,annot=True,fmt='d',cmap='YlGnBu')
```
